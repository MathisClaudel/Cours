1D visualization with Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Create some data
rng = np.random.RandomState(1)
x = np.linspace(0, 10, 50)
z=rng.randn(50, 3)
y = np.cumsum(z, 0)
```

### Plotting Distributions

Apart from the standard histograms <tt>plt.hist</tt>, Seaborn provides smoothed density plots based on data using <tt>sns.kdeplot</tt> or <tt>sns.distplot</tt>.

```python
data = np.random.multivariate_normal([0, 1.5], [[1, 0.2], [0.2, 2]], size=6000)
data = pd.DataFrame(data, columns=['x', 'y'])
```

```python
data.shape
```

```python
data.head()
```

```python
data['x'].min(), data['x'].max()
```

```python
data['y'].min(), data['y'].max()
```

```python
data.mean()
```

```python
data.cov()
```

```python
for col in 'xy':
  print(col)
  plt.hist(data[col], alpha=0.7) # alpha=0.5 provides semi-transparent plots
```

<tt>kdeplot</tt> provides density plots from an array or series (<tt>fill=True</tt> provide filled ones).

```python
sns.kdeplot(data['x'])
sns.kdeplot(data['y'],fill=True)
```

<tt>displot</tt> is a mix of the two previous ones.

```python
sns.displot(data['x'])
sns.displot(data['y'])
```

Joint distribution and the marginal distributions can be displayed together using <tt>jointplot</tt>

```python
sns.jointplot(data=data, x="x", y="y")
```

### Exploring features correlations and interest to classification

Seaborn provides an efficient tool for quickly exploring different features and classification with <tt>pairplot</tt>.

```python
import pandas as pd
import numpy as np

iris = pd.read_csv('/content/iris.csv')
print(iris.shape)
iris.head(2)
```

```python
sns.pairplot(iris, hue='variety')
```

<tt>factorplot</tt> also provides error plots.

```python
sns.boxplot(data=iris, x="sepal.length", y="variety")
```
